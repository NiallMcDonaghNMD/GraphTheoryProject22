<h2>The Heap Data Layout</h2>

As mentioned  in the introduction, Heap is a tree based data structure. For a binary tree to be classified as a heap data structure it must be:
- A complete binary tree (All levels in the tree are filled EXCEPT the lowest possible one)

There are two distinctions within the heap data layout, a Max-Heap and a Min-Heap.

A Max-Heap is when all the children are smaller than their parents. 
E.G. In the above example, 16 is at the top of the tree.

A Min-Heap is the opposite of a Max-Heap, where the children are larger than their parents. 
If using the above example, 1 would be the root of the tree.

<h2>"Heapifying" a tree</h2>

I will demonstrate using code on how to heapify a list of elements, but first I will walk through how to do it manually first, to understand the code better.

If you have an array of elements like [60, 30, 5] there is no need to heapify the list, 
as the parent is already the largest.

If you have an array of elements like [5, 60, 6] there is a need to heapify the list.
so, if you heapified the list you would move 60 to the top of the tree.

The array of elements would now look like [60,5,6].
To make it properly max-heap, you would switch 5 and 6 so the array would look like [60,6,5].

This shows that Heap-Sort is a **Recursive Algorithm**.
Meaning it checks one element to another element, if it is < or > the element being checked, it gets moved up or down accordingly.

<h1>Implementation of heap sort</h1>

````python
# Python program for implementation of heap Sort - original code from https://www.geeksforgeeks.org/python-program-for-heap-sort/

# To heapify subtree rooted at index i.
# sH is size of heap
def heapify(array, sH, i):
	largest = i #initialise largest as root
	left = 2 * i + 1	 # left = 2*i +1, same as theory from above.
	right = 2 * i + 2	 # right = 2*i +2, same as theory from above.

    # See if the left child of root exists and is greater than the root
	if left < sH and array[i] < array[left]:
		largest = left

	#See if the right child of root exists and is greater than the root
	if right < sH and array[largest] < array[right]:
		largest = right

    #Change the root if needed
	if largest != i:
		array[i],array[largest] = array[largest],array[i] # swap

		#Call heapify function to heapify the root
		heapify(array, sH, largest)

# The main function to sort the array of a given size that is passed into it.
def heapSort(array):
	sH = len(array)

	# Build a maxheap. Since last parent will be at ((n//2)-1) we can start at that location.
	for i in range(sH // 2 - 1, -1, -1):
		heapify(array, sH, i)
    # Extract the elements one by one.
	for i in range(sH-1, 0, -1):
		array[i], array[0] = array[0], array[i] 
		heapify(array, i, 0)

````
